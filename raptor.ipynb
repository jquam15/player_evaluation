## Team Data

**I am getting the team salary data from hoopshype https://hoopshype.com/salaries/2021-2022/**

```python
#this function takes a numpy array containing every year in the range that you want and scrapes the hoopshype website to
#get team/player salaries for those years. The function returns a dataframe with every team/player salary from every year in the 
#year_range numpy array. The parameter data_type is a string that takes the value "team" or "players" depending on 
#which data you need
def scrape_salary(year_range, data_type, big_df=pd.DataFrame()):
    for year in year_range:
        if data_type == "team":
            url = f"https://hoopshype.com/salaries/{year-1}-{year}/"
        elif data_type == "players":
            url = f"https://hoopshype.com/salaries/players/{year-1}-{year}/"
        else:
            return "You have entered an incorrect parameter. Pass either 'team' or 'players' as the value for the data_type parameter"
        #The year 2014 corresponds with the 2013-14 season in the url
        
        #print statement for debugging    
        print("Visiting: " + url)
        #get the page html
        page = requests.get(url)
        #check for status exceptions
        page.raise_for_status()
        #get the dataframes on the page
        df_list = pd.read_html(page.text)
        #we want the first one
        year_df = df_list[0]
        #drop the first column as we don't need it (the ranking of which teams have the highest salary)
        year_df = year_df.drop(columns=year_df.columns[0])
        #rename the other columns appropriately
        year_df = year_df.rename(columns={year_df.columns[-2]: "salary", 
                                          year_df.columns[-1]: "inf_adj_salary"})
        #create a year column to specify the year (matched up accordingly with the player raptor years)
        year_df["year"] = year
        #append the df for the year to the final df to be returned
        big_df = big_df.append(year_df)
        #take 2 seconds in between loop iterations to be curteous towards the website
        time.sleep(2)
    
    #the salary values are character types and i want numeric types so I need to use regex to modify these
    #the string values are of the format in the following 6 components:
    #$ (1) numbers (2)  comma (3) numbers (4) comma (5) numbers (6)
    #I only want the numbers so I use the callback \\2\\4\\6 to just get those pieces
    big_df['salary'] = [re.sub("([$])(\\d+)(,)(\\d+)(,*)(\\d*)", "\\2\\4\\6", str(x)) for x in big_df['salary']]
    big_df['inf_adj_salary'] = [re.sub("([$])(\\d+)(,)(\\d+)(,*)(\\d*)", "\\2\\4\\6", str(x)) for x in big_df['inf_adj_salary']]
    #convert salaries into numeric types
    big_df = big_df.astype({"salary": float, "inf_adj_salary": float})
    #return the final df
    return big_df
```

```python
#set a filename
filename = "team_salaries.csv"
#check if the file already exists
if not os.path.exists(os.path.join("data", filename)):
    #take start time
    t1 = time.time()
    #call function with years 2014 to 2022 (the following numpy array has all those years)
    df = scrape_salary(np.arange(2014, 2023), "team")
    #write dataframe to csv file
    df.to_csv(os.path.join("data", filename), index=False)
    #take end time
    t2 = time.time()
    #print how much time it took
    print("File downloaded successfully. It took " + str(t2-t1) + " seconds")
else:
    #if file exists already print that
    print("File already exists")
```

## Win Value (Merging Data)

**The raptor dataset that I used to get the "team_war" dataframe uses team abbreviations where as the hoopshype data uses full city names. In order to merge I need to manually match these up. I do this by mapping a city name to its abbreviation in a dictionary** 

```python
#set up the dictionary
abbr_dict = {"Cleveland": "CLE", "New York": "NYK", "Detroit": "DET", "LA Lakers": "LAL", "Atlanta": "ATL",
            "Dallas": "DAL", "Philadelphia": "PHI", "Milwaukee": "MIL", "Phoenix": "PHO", "Brooklyn": "BRK",
            "Boston": "BOS", "Portland": "POR", "Golden State": "GSW", "San Antonio": "SAS", "Indiana": "IND",
            "Utah": "UTA", "Oklahoma City": "OKC", "Houston": "HOU", "Charlotte": "CHA", "Denver": "DEN",
            "LA Clippers": "LAC", "Chicago": "CHI", "Washington": "WAS", "Sacramento": "SAC", "Miami": "MIA",
            "Minnesota": "MIN", "Orlando": "ORL", "Memphis": "MEM", "Toronto": "TOR", "New Orleans": "NOP"}
len(abbr_dict)
```

```python
#map the full city name to its abbreviation and create the new "team" file
team_salaries["Team"] = team_salaries["Team"].map(abbr_dict)
#inspect data
team_salaries.tail()
```

```python
#rename columns in so that they match up with the raptor data so it is easier to merge the data
team_salaries = team_salaries.rename(columns={"Team": "team", "year": "season"})
#inspect the data
team_salaries.tail()
```

```python
#merge the team salalries with the team war in a new dataframe on the team and season columns
merged_df = team_war.merge(team_salaries, on=["team", "season"])
#inspect the merged data
merged_df.tail()
```

**The next thing I'm going to do is sum all of the team salaries as well as all of the war for each season to put a dollar value on how much 1 raptor win above replacement is worth**

```python
#I want to sum over theese 3 columns
agg_dict = {"war_total": "sum", "team_salaries": "sum", "inf_adj_team_salaries": "sum"}
#group by the season and aggregate (sum) over the 3 columns
win_value = merged_df.groupby("season").agg(agg_dict).reset_index()
#create a dollars_per_war column that assigns a dollar value for 1 win above replacement (WAR) 
win_value["dollars_per_war"] = win_value["team_salaries"] / win_value["war_total"]
win_value["inf_adj_dollars_per_war"] = win_value["inf_adj_team_salaries"] / win_value["war_total"]
#inspect the new df
win_value
```

```python
#create a dictionary with keys as the year and values as dollars per war
win_val_dict = win_value.set_index('season')['dollars_per_war'].to_dict()
#inspect
win_val_dict
```

**This next code cell multiplies the value of a win above replacement (WAR) for the season (according to the win_val_dict that is defined above) by the war total for that season for each row of the dataset.**

**For 2022 for example a win above replacement was worth a little under 5 million dollars**

```python
#set up empty column for player value (war * dollars/war --> unit of dollars)
player_merged["player_val"] = 0.0
#loop through every player in the player_merged dataset
for i in range(len(player_merged)):
    #find which season this is for
    season = player_merged.iloc[i]["season"]
    #determine player value by multiplying the value of war by the players total war to get a unit of dollars
    player_val = win_val_dict[season] * player_merged.iloc[i]["war_total"]
    #add the player value to the row we are on in the loop
    player_merged.iat[i, -1] = player_val 
```

```python
#round the value column to 2 decimals
player_merged["player_val"] = player_merged["player_val"].round(decimals=2)
#drop rows with na values (only 1 row)
player_merged = player_merged.dropna().reset_index(drop=True)
#inspect the data
player_merged.head()
```

**This code block adds a couple columns: one with a 3 year average player value, and one with a long run average player value (data starts in 2013-14 so at most 9 years). Obviously, for players with 3 years or less of service, the 3 year average and long run average will be the same. I'm adding these because they give a more comprehensive sample size to evaluate a player off of as opposed to evaluating them strictly off 1 year**

```python
#execute this code if the file does not exist
filename = "player_value.csv"
if not os.path.exists(os.path.join("data", filename)):
    #create columns
    player_merged["3_year_average_val"] = 0.0
    player_merged["lr_average_val"] = 0.0
    #get start time
    t1 = time.time()
    for i in range(len(player_merged)):
        #get the players name for the row
        name = player_merged.iloc[i]["player_name"]
        #get the season for the row
        season = player_merged.iloc[i]["season"]
        #for the 3 year value, filter the player_merged dataframe to only the rows with the current name value as well as the
        #the current season along with the prior 2 seasons...average the player val over those 3 seasons
        #additionally if the player has not player 3 seasons this dataframe will be filtered to only those rows where
        #the name is present...so if the name is present only once, then the dataframe will be one row (won't throw error)
        #some players will have been traded and will have more than 3 rows...so we want to sum over the values in the 3 
        #year df and divide by number of unique seasons (3 if they've played 3 more more years otherwise 2 or 1)
        three_year_val = player_merged[(player_merged["player_name"] == name) & (player_merged["season"] <= season) & 
                               (player_merged["season"] >= season-2)]["player_val"].sum()
        seasons = len(player_merged[(player_merged["player_name"] == name) & (player_merged["season"] <= season) & 
                             (player_merged["season"] >= season - 2)]["season"].unique())
        #for the career value, filter the player_merged dataframe to only the rows with the current name except this time average
        #over every season up to that point in their career
        #***DO NOT ADD EVERY SEASON (ONLY THOSE PRIOR) SINCE IN REAL LIFE WE CAN'T PREDICT FUTURE SEASONS***
        #in the case of players being traded, sum and divide by number of unique years 
        lr_val = player_merged[(player_merged["player_name"] == name) & (player_merged["season"] <= season)]["player_val"].sum()
        num_seasons = len(player_merged[(player_merged["player_name"] == name) & (player_merged["season"] <= season)]["season"].unique())
        #add 3 year and career averages
        player_merged.iat[i, -2] = three_year_val / seasons
        player_merged.iat[i, -1] = lr_val / num_seasons
    #get end time
    t2 = time.time()
    #round df to only 2 decimal values
    player_merged = player_merged.round(2)
    #download dataframe to a csv
    player_merged.to_csv(os.path.join("data", filename), index=False)
    print(f"File name {filename} downloaded")
else:
    print(f"File named {filename} already exists")
```

## Overvalued and Undervalued Players

**To determine the disparity between a player's and their contract value I will take the player's 3 year average value (because a 3 year sample size is more reliable) and subtract the player's contract value. As a result, players with positive values in the "diff" column have been more valuable than what their contract indicates while players with negative values in the "diff" column  have been less valuable than what their contract indicates**

```python
#create the difference column
pv["diff"] = pv["3_year_average_val"] - pv["salary"]
#I'm going to filter the data so that only players with >=820 minutes stay in (10 minutes per game for 82 game season)
pv = pv[pv["mp"] >= 820].reset_index(drop=True)
#inspect the data
pv.head()
```

```python
#create an undervalued df sorted by diff in descending order
undervalued = pv.sort_values("diff", ascending=False).round(2)
#look at the top 20 most undervalued players
undervalued.head(20)
```

```python
#create an overvalued df sorted by diff in ascending order
overvalued = pv.sort_values("diff").round(2)
#look at the top 20 most overvalued players
overvalued.head(20)
```
